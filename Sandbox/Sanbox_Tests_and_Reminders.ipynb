```
# Shopping Cart Program

import pandas as pd

print("Welcome to the shopping cart program")

n_items = int(input("How many items do you want to add to the cart?: "))

items = []
prices = []
quantity = []

total_price = 0

for i in range(n_items):
  item_name = str(input(f"What is the item {i+1}?: "))
  item_price = float(input(f"What is the price of item {i+1}?: "))
  item_quantity = int(input(f"How many {item_name} do you want to buy?: "))
  items.append(item_name)
  prices.append(item_price)
  quantity.append(item_quantity)


# The print statements were also trying to print generator objects directly.
# To print the items and prices in a readable format, iterate through them or print the lists directly.
# Here, we iterate to print each item and its corresponding price.
print("\n--- Your Cart ---")
for i in range(n_items):
    print(f"{items[i]}: ${prices[i]} {quantity[i]:.2f}")

df = pd.DataFrame({'Item': items, 'Price': prices, 'Quantity': quantity})
print("\n",df)

total_price = sum(df['Price']*df['Quantity'])
print(f"\nTotal price: ${total_price:.2f}")
```
